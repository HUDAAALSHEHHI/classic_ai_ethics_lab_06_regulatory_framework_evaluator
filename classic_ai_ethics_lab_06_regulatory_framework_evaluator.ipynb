```python
import pandas as pd

data = pd.DataFrame({
    "Policy_ID": [1, 2, 3, 4],
    "Has_AI_Compliance": [True, False, True, True],
    "Legal_Risk_Score": [0.2, 0.7, 0.5, 0.3],
    "Governance_Clarity": [0.9, 0.4, 0.6, 0.8]
})
data
```

```python
def evaluate_policy(row):
    if not row["Has_AI_Compliance"]:
        return "Violation: Missing AI compliance"
    if row["Legal_Risk_Score"] > 0.6:
        return "Alert: High legal risk"
    if row["Governance_Clarity"] < 0.5:
        return "Warning: Low governance clarity"
    return "Compliant"

data["Policy_Status"] = data.apply(evaluate_policy, axis=1)
data
```

```python
issues = data[data["Policy_Status"] != "Compliant"]
print("Detected compliance issues:", len(issues))
print("\nDetailed Report:\n", data)
```
